### 10-4
- 完全平行二分木の議論より, ノードの数をN, 木の高さをhとした時, 強平行二分木は以下のような不等式が成り立つ
    - $$2^h \leq N \leq 2^{h+1} - 1$$
- よって, $\log(N) - 1 \leq h \leq \log(N)$ より$h = \log(N)$ 

### 10-5
数学的帰納法で示す.
- $N = 1$の時
辺は存在しないため, $|E| = N-1$が成り立つ.

- 木の頂点数が$N \leq m$の時, $|E| = N-1$が成り立つと仮定する.
    - 任意に$N = m+1$である木$T$を固定する.
    -  $T$は木であることから次数が1以下である頂点をもつ
        - $T$の最長パスを$P$とする.このとき, $u \in P$が葉でないとするとvと隣接する頂点$v, w$が存在する.
        - $T$は木であるから, $v, w$のうち少なくとも一つは$P$に含まれていない必要がある. その頂点をvとする.
        - このとき, $P$に$v$を加えた$P'$は$P$より長くかつパスである.よって矛盾.
    - 次数1の頂点とその辺を取り除いた$T'$は構成より木でありかつ頂点数は$m$
    - よって, 仮定より$T'$の辺の数は$m-1$であるから, $T$の辺の数は$m$
