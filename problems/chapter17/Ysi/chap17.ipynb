# 17-1
### 問
- ハミルトンサイクル問題がNPに属することを示せ

### 答
- ハミルトンサイクルの解の候補が与えられた時、これは問題の入力よりも短い長さで書き表すことができる
- また、与えられた解候補のパスを一辺ずつ見ていき、Vの要素のみで全ての要素が利用されていること、Eの要素のみで高々一度までしか利用されていないことを判定していく。
- この確認は多項式時間で実施可能なめ、ハミルトンサイクル問題は多項式検証機を持
- よってハミルトンサイクル問題はクラスNPに属する

# 17-2
### 問
- クリーク問題がNPに属することを示せ
### 答
- クリーク問題の解の候補となる部分グラフが与えられた時、これは問題の長さよりも短い長さで書き表すことができる。
- また与えられた解候補の、頂点i, j間の辺がEに存在しているかを確認する。
- この確認は多項式時間で可能なため、クリーク問題は多項式検証機を持つ
- よって、クリーク問題はクラスNPに属する



# 17-3
### 問
- 安定集合問題がNP完全であることを用いてクリーク問題がNP完全であることを示せ
### 答
- クリーク問題のインスタンス$G = (V, E), k$ に対して、 $E$において辺が存在する場合には辺除き、辺が存在しない場合には辺を加えた$\bar{E}$ を考える.
- $E$から$\bar{E}$ を構成するのは多項式時間で可能.
- ここで、$ \bar{G} = (V, \bar{E}), k $に対して安定集合問題の解を考えると,これはGにおけるクリーク問題の解である頂点集合である.
- よって, 安定集合問題$\leq_{p}$クリーク問題
- 安定集合問題がNP完全であるから, クリーク問題もNP完全問題.
